### Dsitribution of the Data

```python
numeric_cols = df.select_dtypes(include=np.number).columns

df[numeric_cols].hist(figsize=(15, 12), bins=30)
plt.suptitle("Distributions of Numeric Features")
plt.tight_layout()
plt.show()
```

#### Monthly Sales Trends 


```python
df['year_month'] = pd.to_datetime(df['year_month'], errors='coerce', dayfirst=True)
sales_trend = df.groupby('year_month')['new_policy_count'].sum().reset_index()

plt.figure(figsize=(12, 5))
plt.plot(sales_trend['year_month'], sales_trend['new_policy_count'], marker='o')
plt.title("Total Policies Sold per Month")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

#### Multivariate Analysis

```python
plt.figure(figsize=(12, 8))
corr_matrix = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Matrix')
plt.show()
```

### PCA for Multivariate Analysis

```python
scaled_data = StandardScaler().fit_transform(df[numeric_cols].dropna())
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

plt.figure(figsize=(10, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], alpha=0.5)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("PCA of Numeric Features")
plt.grid()
plt.show()
```

### PCA Results Interpretation

The PCA scatter plot shows that most agents' data points are clustered near the center. This indicates that, in terms of the top two principal components (which capture the largest variance in the dataset), the majority of agents exhibit similar behavior and feature values.

There is no strong clustering pattern, suggesting low separability in the top two components. This may mean agent performance is not easily separable based on raw features alone, reinforcing the need for engineered features (e.g., engagement score, lag-based trends) or advanced modeling techniques to reveal hidden patterns.


#### Individual Agent Growth over time (Agent Trajectory) 

```python
sample_agents = df['agent_code'].value_counts().index[:5]  # Top 5 active agents

plt.figure(figsize=(12, 6))
for agent in sample_agents:
    agent_data = df[df['agent_code'] == agent].sort_values('year_month')
    plt.plot(agent_data['year_month'], agent_data['new_policy_count'], label=f"Agent {agent}")

plt.title("Agent Sales Trajectories")
plt.xlabel("Month")
plt.ylabel("New Policies Sold")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

#### Agent Trajectories Over Time

This line plot tracks the monthly policy sales for the top 5 performing agents. It reveals patterns such as:

- **Consistency**: Some agents show steady sales, suggesting reliable performance.
- **Volatility**: Others have fluctuating trends, possibly due to external factors or inconsistent effort.
- **Drop-offs**: Agents whose sales decline over time may need motivation.

Analyzing such trends helps identify early signs of underperformance and introduce strategies to improve their performance.


#### Engagement Score

```python
df['engagement_score'] = (
    df['unique_customers_last_7_days'] +
    df['unique_quotations_last_7_days'] +
    df['unique_proposals_last_7_days']
)

plt.figure(figsize=(12, 5))
sns.scatterplot(data=df, x='engagement_score', y='new_policy_count')
plt.title("Engagement Score vs New Policy Count")
plt.grid()
plt.show()
```

### Engagement Score vs Policy Sales

We created a simple "engagement score" to quantify agent activity in the most recent week. This score combines the number of customers contacted, quotations issued, and proposals submitted.

The scatter plot suggests a positive relationship â€” agents with higher engagement scores tend to sell more policies. This reinforces the importance of consistent customer interaction and proposal generation for driving sales.

However, there are still agents with high engagement but low sales, hinting at possible issues with lead quality, conversion skills, or follow-up practices. These agents could benefit from targeted training or mentoring.
